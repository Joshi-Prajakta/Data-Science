```python
# Open the file in read only mode.
text_file= open('D:/DATASCIENCE/Python/Project/conv.txt',"r",encoding='utf-8')
```

```python
# read the file and store the text in a variable 'complete_text.'
complete_text=text_file.read()
complete_text
```

- ### <B><font color=black>Cleaning the Text File:</font>
---

```python
#create a variable "special_characters" to store the puctuation marks.
special_characters = '''….?!,''' 

for element in complete_text:
    if element in special_characters:
        complete_text = complete_text.replace(element, "") #This removes the punctuation marks from complete_text.
        
print(complete_text)
```

```python
#create a variable "replacements" to store the abbreviation with their full form respectively.
replacements=[("I’ve", "I have"),("he’ll", "he will"),("It’s", "It is"),(" WILL", "WILL"),("Father’s", "Father is"),("you’ve", "you have"),("They’ve", "They have"),("he’s", "he is"),("mother’s", "mother is"),("we’re", "we are"),("I’m", "I am"),("won’t", "will not"),("Don’t", "Do not"),("don’t", "do not"),("didn’t", "did not"),("That’s", "That is"),("You’re", "You are"),("haven’t", "have not"),("Where’s", "Where is"),("We’ve", "We have"),("You’ll", "You will"),("He’s", "He is"),("You’ve", "You have"),("16", "sixteen")]

for char, replacement in replacements:
    if char in complete_text:
        complete_text= complete_text.replace(char,replacement) #This replaces the abbreviation from complete_text to respective full form.

print(complete_text)
```

```python
script = complete_text.splitlines() #This splits the complete_text(type:string) into a script(type:list) where each line is a list item.

print(script)
```

```python
for line in script:
    script.remove('')  #This removes multiple values of '' from script(type:list).
    continue             
```

###  Task 1:   
- #### The number of unique characters in the sample conversation :


```python
#create lists 'names' and 'speech' with no elements to store the name of the character and sentence spoken by the character with their name.
names  = []
speech = []

for lines in script:
    sentence = lines.split(':') #This splits lines in script to seperate the name of the character from their respective conversations.
    speech.append(sentence) #This stores the sentence with the character name and sentence spoken by the character.
    names.append(sentence[0]) #This stores the names of the character.
```

```python
#create a set 'unique_characters' to store the unique values in 'names'(type:list)
unique_characters=set(names)

print("The Number of unique characters in the sample conversation is ", len(unique_characters)) #len(unique_characters) function gives the number of items in 'unique_characters'.
print( "\nTheir names are as follows : ")

#create a 'for loop' to get the names of the characters with index number starting from 1.
for index, val in enumerate(list(unique_characters), start=1):
    print(index,val)                                            
```


###  Task 2:

- #### Create a new text file for each of the characters with their name and store the unique words said by them in their respective file.
- #### Store one word in one line :

```python
for character in unique_characters:          #This extracts details for each character present in the unique_characters(type:set).
    dialog=[]
    unique_words={}
    for person, i in speech:                 #This stores every conversation of each person in speech(type:list).
            if character == person:
                dialog+=(i.lower()).split()  #'lower()' returns i(type:string) where all characters are lower case.
            unique_words=set(dialog)         #Set returns the unique values in 'dialog' stored as 'unique_words'.    
    
    
#Create a new text file for each character with name and store the unique_words in their respective file.
    with open(character+'.txt',"w") as f:
        for words in unique_words:
            f.write(words+'\n')
```
